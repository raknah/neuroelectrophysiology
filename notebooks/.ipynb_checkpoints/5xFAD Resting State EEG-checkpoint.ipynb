### II.I PSD Analysis

```python
from scipy.signal.windows import dpss
from scipy.fft import rfftfreq, rfft
import numpy as np

def bandpower(trial, bands=None, nw=2):
    if bands is None:
        bands = {
            'delta': (1, 4),
            'theta': (4, 8),
            'alpha': (8, 12),
            'beta': (12, 25),
            'gamma': (25, 50)
        }

    fs = trial.sampling_rate
    epochs, channels, samples = trial.data.shape
    frequencies = rfftfreq(samples, d=1/fs)

    tapers = dpss(samples, NW=nw, Kmax=2*nw - 1)
    k = tapers.shape[0]

    psd = np.zeros((epochs, channels, len(frequencies)))
    features = np.zeros((epochs, channels, len(bands)))

    for epoch in range(epochs):
        for channel in range(channels):
            signal = trial.data[epoch, channel, :]
            spectrum_sum = np.zeros_like(frequencies)

            for taper in tapers:
                tapered = signal * taper
                spectrum = np.abs(rfft(tapered)) ** 2
                spectrum_sum += spectrum

            psd[epoch, channel] = spectrum_sum / k

            for idx, (band_name, (low, high)) in enumerate(bands.items()):
                mask = (frequencies >= low) & (frequencies <= high)
                features[epoch, channel, idx] = np.sum(psd[epoch, channel, mask])

    trial.data = features  # shape: (epochs, channels, bands)
    return trial
```

```python
from openephysextract.utilities import savify

features = [bandpower(trial) for trial in processed]

savify(features, output_2, 'features')
```

### II.II Logistic Scaling (Garwood et. al)

```python
def scaler(trial):
    
    features = trial.data
    scaled = np.zeros_like(features)
    
    for i in range(trial.data.shape[1]):
        x = features[:, i]
        q1 = np.percentile(x, 25)
        q3 = np.percentile(x, 75)
        median = np.median(x)
        IQR = q3 - q1
        
        if IQR == 0:
            scaled[:, i] = (x - np.min(x))/(np.max(x) - np.min(x))
        else:
            lam = (2 * np.log(3)) / IQR
            scaled[:, i] = 1 / (1 + np.exp(-lam * (x - median)))\
            
    trial.data = scaled
    
    return trial
```

```python
scaled = [scaler(trial) for trial in features]

savify(scaled, output_2, "logistic-scaled")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Button, TextBox
from matplotlib import gridspec
import seaborn as sns

class Viewer:
    def __init__(self, trials, band_names=None):
        self.trials = trials
        self.n_trials = len(trials)
        self.band_names = band_names or ['delta', 'theta', 'alpha', 'beta', 'gamma']
        self.current_index = 0

        # Preallocate figure and axes
        self.fig = plt.figure(figsize=(21, 10))
        gs = gridspec.GridSpec(2, 2, width_ratios=[10, 1], height_ratios=[2, 1])
        plt.subplots_adjust(bottom=0.25, hspace=0.4, wspace=0.3)

        self.ax_line = self.fig.add_subplot(gs[0, 0])
        self.ax_legend = self.fig.add_subplot(gs[0, 1])
        self.ax_heatmap = self.fig.add_subplot(gs[1, 0])
        self.ax_cbar = self.fig.add_subplot(gs[1, 1])
        self.ax_legend.axis('off')

        # Widgets
        self.ax_prev = plt.axes([0.1, 0.05, 0.1, 0.075])
        self.ax_next = plt.axes([0.21, 0.05, 0.1, 0.075])
        self.ax_box  = plt.axes([0.4, 0.06, 0.2, 0.05])

        self.btn_prev = Button(self.ax_prev, 'Prev')
        self.btn_next = Button(self.ax_next, 'Next')
        self.box_jump = TextBox(self.ax_box, 'Go to Trial')

        self.btn_prev.on_clicked(self.prev_trial)
        self.btn_next.on_clicked(self.next_trial)
        self.box_jump.on_submit(self.goto_trial)

        # Connect zoom/scroll
        self.fig.canvas.mpl_connect('scroll_event', self.on_scroll)

    def get_avg_band_power(self, trial):
        return trial.data.mean(axis=1)  # (epochs, bands)

    def update_plot(self):
        trial = self.trials[self.current_index]
        genotype = trial.notes.get('genotype', 'Unknown')
        avg_power = self.get_avg_band_power(trial)

        self.ax_line.clear()
        lines = []
        for b, band in enumerate(self.band_names):
            line, = self.ax_line.plot(avg_power[:, b], label=band)
            lines.append(line)

        self.ax_line.set_title(f"Trial {self.current_index} — Genotype: {genotype}")
        self.ax_line.set_xlabel("Epoch")
        self.ax_line.set_ylabel("Power")

        self.ax_legend.clear()
        self.ax_legend.axis('off')
        self.ax_legend.legend(lines, self.band_names, loc='center')

        self.ax_heatmap.clear()
        self.ax_cbar.clear()

        sns.heatmap(avg_power.T, ax=self.ax_heatmap, cbar_ax=self.ax_cbar,
                    cmap="viridis", yticklabels=self.band_names)
        self.ax_heatmap.set_title("Band Power Heatmap")
        self.ax_heatmap.set_xlabel("Epoch")
        self.ax_heatmap.set_ylabel("Band")

        self.ax_line.relim()
        self.ax_line.autoscale_view()
        self.ax_heatmap.relim()
        self.ax_heatmap.autoscale_view()

        self.fig.canvas.draw_idle()

    def prev_trial(self, event):
        if self.current_index > 0:
            self.current_index -= 1
            self.update_plot()

    def next_trial(self, event):
        if self.current_index < self.n_trials - 1:
            self.current_index += 1
            self.update_plot()

    def goto_trial(self, text):
        try:
            index = int(text)
            if 0 <= index < self.n_trials:
                self.current_index = index
                self.update_plot()
        except ValueError:
            pass

    def on_scroll(self, event):
        """Scroll to zoom horizontally on line plot only."""
        if event.inaxes != self.ax_line:
            return

        ax = self.ax_line
        xlim = ax.get_xlim()
        x_range = xlim[1] - xlim[0]
        x_mouse = event.xdata

        # Zoom factor: <1 = zoom in, >1 = zoom out
        scale_factor = 1.2 if event.button == 'up' else 1 / 1.2

        new_left = x_mouse - (x_mouse - xlim[0]) * scale_factor
        new_right = x_mouse + (xlim[1] - x_mouse) * scale_factor

        ax.set_xlim([new_left, new_right])
        self.fig.canvas.draw_idle()

    def view(self):
        self.update_plot()
```

```python


viewer = Viewer(scaled)
viewer.view()
```
